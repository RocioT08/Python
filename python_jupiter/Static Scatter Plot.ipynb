```python
import seaborn as sns  
import scipy as sci
import ssl
ssl._create_default_https_context = ssl._create_unverified_context

mpg_df = sns.load_dataset("mpg") 
ax = sns.scatterplot(x="weight", y="mpg", data= mpg_df)
```

```python
#HexBin Plots
## set the plot style to include ticks on the axes.
sns.set(style="ticks")
## hexbin plot
sns.jointplot(x=mpg_df.weight, y=mpg_df.mpg, kind="hex", color="#4CB391")
```

```python
#Ex3: Static Contour Plots
mpg_df = sns.load_dataset("mpg")
sns.set_style("white")
#Generate a Kernel Density Estimate (KDE) 
# generate KDE plot: first two parameters are arrays of X and Y coordinates of data points 
# parameter shade is set to True so that the contours are filled with a color gradient based on number of data points
sns.kdeplot(x=mpg_df.weight, y= mpg_df.mpg, Fill=True)
```

```python
#Ex4: Static Line Plot

ax1 = sns.scatterplot(x="model_year", y="mpg", data=mpg_df)
```

```python
#Draw a simple line plot to show the relationship between model_year and mileage
ax= sns.lineplot(x="model_year", y="mpg", data=mpg_df)
```

```python
sns.lineplot(x="model_year", y="mpg", data=mpg_df, ci=68)
```

```python
#Ex5: Multiple Line Plots  Data across Time
flights_df = sns.load_dataset("flights")
print(flights_df.head())

#Create multiple plots for the months of December and January:
ax= sns.lineplot(x="year", y= "passengers", data=flights_df[flights_df['month']=='Jan'], color='green')
ax= sns.lineplot(x="year", y= "passengers", data=flights_df[flights_df['month']=='Feb'], color='blue')
ax= sns.lineplot(x="year", y= "passengers", data=flights_df[flights_df['month']=='Mar'], color='pink')
ax= sns.lineplot(x="year", y= "passengers", data=flights_df[flights_df['month']=='Apr'], color='yellow')
ax= sns.lineplot(x="year", y= "passengers", data=flights_df[flights_df['month']=='May'], color='orange')
ax= sns.lineplot(x="year", y= "passengers", data=flights_df[flights_df['month']=='Jun'], color='brown')
ax= sns.lineplot(x="year", y= "passengers", data=flights_df[flights_df['month']=='Jul'], color='red')
ax= sns.lineplot(x="year", y= "passengers", data=flights_df[flights_df['month']=='Aug'], color='black')
ax= sns.lineplot(x="year", y= "passengers", data=flights_df[flights_df['month']=='Sep'], color='purple')
ax= sns.lineplot(x="year", y= "passengers", data=flights_df[flights_df['month']=='Oct'], color='gold')
ax= sns.lineplot(x="year", y= "passengers", data=flights_df[flights_df['month']=='Nov'], color='beige')
ax= sns.lineplot(x="year", y= "passengers", data=flights_df[flights_df['month']=='Dec'], color='gray')
```

```python
#Heat map
df_pivoted= flights_df.pivot(index='month',columns='year',values='passengers')
ax= sns.heatmap(df_pivoted)
```

```python
#ClusterMap

ax=sns.clustermap(df_pivoted, col_cluster = False, row_cluster=True)
```
