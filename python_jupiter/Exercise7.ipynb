```python
import pandas as pd
import plotly.express as px

import ssl
ssl._create_default_https_context = ssl._create_unverified_context


gm_url= "https://raw.githubusercontent.com/TrainingByPackt/Interactive-Data-Visualization-with-Python/master/datasets/gapminder.csv"
gm= pd.read_csv(gm_url)

co2_url= "https://raw.githubusercontent.com/TrainingByPackt/Interactive-Data-Visualization-with-Python/master/datasets/co2.csv"
co2= pd.read_csv(co2_url)

#gm = sns.load_dataset("gapminder")
gm = pd.read_csv(gm_url)
co2 = pd.read_csv(co2_url)
co2.head()

df_gm = gm[['Country', 'region']].drop_duplicates() 
df_w_regions = pd.merge(co2, df_gm, left_on='country', right_on='Country', how='inner') 
df_w_regions = df_w_regions.drop('Country', axis='columns') 
new_co2 = pd.melt(df_w_regions, id_vars=['country', 'region']) 
columns = ['country', 'region', 'year', 'co2'] 
new_co2.columns = columns 
df_co2 = new_co2[new_co2['year'].astype('int64') > 1963] 
df_co2 = df_co2.sort_values(by=['country', 'year']) 
df_co2['year'] = df_co2['year'].astype('int64') 
df_gdp = gm[['Country', 'Year', 'gdp']] 
df_gdp.columns = ['country', 'year', 'gdp'] 
data = pd.merge(df_co2, df_gdp, on=['country', 'year'], how='left') 
data = data.dropna() 
data.head()
```

```python
# Create a scatter plot with box plot for co2 values (marginal y)
scatter_box = px.scatter(data, x='year', y='co2', marginal_y='box')
scatter_box.show()

# Create a rug plot for gdp values (marginal x) with animation on the year column
rug_animation = px.scatter(data, x='gdp', y='co2', animation_frame='year', marginal_x='rug')
rug_animation.show()

# Create a scatter plot of gdp vs co2
scatter_gdp_co2 = px.scatter(data, x='gdp', y='co2')
scatter_gdp_co2.show()

# Create a density contour of gdp vs co2
density_contour = px.density_contour(data, x='gdp', y='co2')
density_contour.show()
```
